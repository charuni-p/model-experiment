```python
import scipy.stats
from scipy import signal
import xarray as xr
```

```python
ds_era5 = xr.open_mfdataset('/g/data/e14/cp3790/Charuni/NCI/ERA5-T2M-MEAN/t2m_anom_noaa_remap.nc',
                           chunks = {'lon':120, 'lat':80, 'time':90}).sel(time=slice('1982', '2018'))
sata = ds_era5.t2m_anomaly.sel(lat=slice(-53, -30), lon=slice(135, 180))
```

```python
sata_mon = sata.resample(time='1M').mean(dim='time')
```

```python
# apply_ufunc call
test_detrend = xr.apply_ufunc(scipy.signal.detrend,sata_mon.dropna(dim='time'),kwargs=dict(axis=2),
                              input_core_dims=[['time']],output_core_dims=[['time']],
                              dask='allowed',vectorize=True)
```

```python
test_detrend
```

```python
ds_gm = xr.open_mfdataset(['/g/data/e14/asg561/MHW_NOAAupdated/mhw_data_90pc/mhw_severity.pc90.120to150.-50to-30.1981.2021.nc',
                           '/g/data/e14/asg561/MHW_NOAAupdated/mhw_data_90pc/mhw_severity.pc90.120to150.-70to-50.1981.2021.nc',
                           '/g/data/e14/asg561/MHW_NOAAupdated/mhw_data_90pc/mhw_severity.pc90.150to180.-50to-30.1981.2021.nc',
                           '/g/data/e14/asg561/MHW_NOAAupdated/mhw_data_90pc/mhw_severity.pc90.150to180.-70to-50.1981.2021.nc'],
                         chunks = {'longitude':120, 'latitude':80, 'time':90}).sel(time=slice('1982', '2018'))
```

```python
ssta = ds_gm.ssta.sel(lat=slice(-53, -30), lon=slice(135, 180))
ssta_mon = ssta.resample(time='1M').mean(dim='time')
```

```python
test_detrend_ssta = xr.apply_ufunc(scipy.signal.detrend,ssta_mon.dropna(dim='time'),kwargs=dict(axis=2),
                              input_core_dims=[['time']],output_core_dims=[['time']],
                              dask='allowed',vectorize=True) 
```

```python
test_detrend_ssta
```

```python
# Use .groupby('time.month') to organize the data into months
# then use .groups to extract the indices for each month
month_idxs=test_detrend.groupby('time.month').groups

# Extract the time indices corresponding to all the Januarys 
jan_idxs=month_idxs[1]

# Extract the january months by selecting 
# the relevant indices
sata_jan=test_detrend.isel(time=jan_idxs)
```

```python
# Use .groupby('time.month') to organize the data into months
# then use .groups to extract the indices for each month
month_idxs=test_detrend_ssta.groupby('time.month').groups

# Extract the time indices corresponding to all the Januarys 
jan_idxs=month_idxs[1]

# Extract the january months by selecting 
# the relevant indices
ssta_jan=test_detrend_ssta.isel(time=jan_idxs)
```

```python
corr_jan = xr.corr(ssta_jan, sata_jan, dim='time')
```
